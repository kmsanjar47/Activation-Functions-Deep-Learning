```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)
```

```python
X = np.linspace(-3, 3, 100)
```

```python
y_sigmoid = sigmoid(X)
y_tanh = tanh(X)
y_relu = relu(X)
```

```python
plt.plot(X, y_sigmoid)
plt.grid(True)
```

```python
plt.plot(X, y_tanh)
plt.grid(True)
```

```python
plt.plot(X, y_relu)
plt.grid(True)
```
